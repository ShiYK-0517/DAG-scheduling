```python
#CPU Case


import numpy as np

G=6
S=3
T=3
C=1




# task type 0: POTRF 1:SYRK 2:TRSM 3: GEMMS
# all have 4 values, but the last one is used for GEMMS only (-;
# a task is a 4-uple and everything will be stored in dictionnaries later
# succalap is the set of successors in the alap schedule, ie the predecessors in the original graph



def succASAP(x,n):
    tasktype=x[0]
    i=x[1]
    j=x[2]
    k=x[3]
    listsucc=[]
    if tasktype==0:
        if i<n:
            for j in range(i+1,n+1,1):
                y=(2,i,j,0)
                listsucc.append(y)
        else:
            y=(4,0,0,0)
            listsucc.append(y)  
    
    
    if tasktype==1:
        if j<i-1:
            y=(1,i,j+1,0)
            listsucc.append(y)
        else:
            y=(0,i,0,0)
            listsucc.append(y)

           
            
    if tasktype==2:
        if i<=n-1:
            for k in range(i+1,j):
                y=(3,k,j,i)
                listsucc.append(y)
            for k in range(j+1,n+1):
                y=(3,j,k,i)
                listsucc.append(y)
            y=(1,j,i,0)
            listsucc.append(y)


    if tasktype==3:
        if k<i-1:
            y=(3,i,j,k+1)
            listsucc.append(y)
        else:
            y=(2,i,j,0)
            listsucc.append(y)


    return listsucc


# duration is just the duration of a task

def duration(x):
    tasktype=x[0]
    if tasktype==0:
        duration=C
    if tasktype==1:
        duration=S        
    if tasktype==2:
        duration=T
    if tasktype==3:
        duration=G
    if tasktype==4:
        duration=0
    return duration

durations = [C, S, T, G]


def CPAndWorkBelow(x,n):
    ReadyTasks=[]
    ReadyTasks.append(x)
    Seen=[]
    ToVisit=[]
    ToVisit.append(x)
    TotalWork=duration(x)
    CPl=0
    while len(ToVisit)>0:
        for t in ToVisit:
            for succ in succASAP(t,n):
                if succ not in Seen:
                    TotalWork=TotalWork+duration(succ)
                    Seen.append(succ)
                    ToVisit.append(succ)
            ToVisit.remove(t)
    CP=2*C + T + S + (n - 2)*(T + G)
    tasktype=x[0]
    if tasktype==0:
        CPl=CP-(x[1]-1)(T +G)
    if tasktype==1:
        if x[1]<n:
            CPl=CP-(x[1]-1)(T +G)+(x[1]-x[2])*S
        else:
            CPl=(n-x[2])*S+C
    if tasktype==2:
        CPl=CP-C-(x[1]-1)*(T+G)
    if tasktype==3:
        CPl=CP-C+G+T -x[1]*T -x[3]*G
    
    
    
    return(CPl,TotalWork)

            

```

```python
 




# GPU Case


import numpy as np

G=3
S=2
T=8
C=11




# task type 0: POTRF 1:SYRK 2:TRSM 3: GEMMS
# all have 4 values, but the last one is used for GEMMS only (-;
# a task is a 4-uple and everything will be stored in dictionnaries later
# succalap is the set of successors in the alap schedule, ie the predecessors in the original graph



def succASAP(x,n):
    tasktype=x[0]
    i=x[1]
    j=x[2]
    k=x[3]
    listsucc=[]
    if tasktype==0:
        if i<n:
            for j in range(i+1,n+1,1):
                y=(2,i,j,0)
                listsucc.append(y)
        else:
            y=(4,0,0,0)
            listsucc.append(y)  
    
    
    if tasktype==1:
        if j<i-1:
            y=(1,i,j+1,0)
            listsucc.append(y)
        else:
            y=(0,i,0,0)
            listsucc.append(y)

           
            
    if tasktype==2:
        if i<=n-1:
            for k in range(i+1,j):
                y=(3,k,j,i)
                listsucc.append(y)
            for k in range(j+1,n+1):
                y=(3,j,k,i)
                listsucc.append(y)
            y=(1,j,i,0)
            listsucc.append(y)


    if tasktype==3:
        if k<i-1:
            y=(3,i,j,k+1)
            listsucc.append(y)
        else:
            y=(2,i,j,0)
            listsucc.append(y)


    return listsucc


# duration is just the duration of a task

def duration(x):
    tasktype=x[0]
    if tasktype==0:
        duration=C
    if tasktype==1:
        duration=S        
    if tasktype==2:
        duration=T
    if tasktype==3:
        duration=G
    if tasktype==4:
        duration=0
    return duration

durations = [C, S, T, G]


def CPAndWorkBelow(x,n):
    ReadyTasks=[]
    ReadyTasks.append(x)
    Seen=[]
    ToVisit=[]
    ToVisit.append(x)
    TotalWork=duration(x)
    CPl=0
    while len(ToVisit)>0:
        for t in ToVisit:
            for succ in succASAP(t,n):
                if succ not in Seen:
                    TotalWork=TotalWork+duration(succ)
                    Seen.append(succ)
                    ToVisit.append(succ)
            ToVisit.remove(t)
        
    tasktype=x[0]
    if tasktype==0:
        CPl=C+(n-x[1])*(T+S+C)
    if tasktype==1:
        CPl=(x[1]-x[2])*S+C+(n-x[1])*(T+S+C)       
    if tasktype==2:
        CPl=(x[2]-x[1]-1)*(T+G)+(n-x[2]+1)*(T+S+C)
    if tasktype==3:
        CPl=(x[1]-x[3])*G+(x[2]-x[1]-1)*(T + G)+(n-x[2] +1)*(T +S +C)    
    
    
    
    return(CPl,TotalWork)

            

    
```
